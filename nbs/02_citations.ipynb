````python
#| eval: false
#| hide
import datasets

mms_dataset = datasets.load_dataset("Brand24/mms")
mms_dataset_df = mms_dataset["train"].to_pandas()
````

````python
#| eval: false
#| hide
from mms_benchmark.citations import get_citations
citations = get_citations(mms_dataset["train"], citation_as_dict=False)
````

````python
#| eval: false
#| hide
mms_dataset_df = mms_dataset_df.drop(columns=["label", "_id", "text", "cleanlab_self_confidence"]).drop_duplicates()
````

## Citations

````python
#| eval: false
#| echo: false
markdown_output = ""

# Iterate over each row in the DataFrame
for index, row in mms_dataset_df.iterrows():
    original_dataset = row['original_dataset']
    domain = row['domain']
    language = row['language']
    family = row['Family']
    genus = row['Genus']
    definite_articles = row['Definite articles']
    indefinite_articles = row['Indefinite articles']
    number_of_cases = row['Number of cases']
    word_order = row['Order of subject, object, verb']
    negative_morphemes = row['Negative morphemes']
    polar_questions = row['Polar questions']
    position_of_negative_word = row['Position of negative word wrt SOV']
    prefixing_vs_suffixing = row['Prefixing vs suffixing']
    coding_of_nominal_plurality = row['Coding of nominal plurality']
    grammatical_genders = row['Grammatical genders']
    
    # Generate markdown for the current row
    markdown_dataset = f"""
### Dataset id: {original_dataset} 

- Domain: **{domain}**
- Language: **{language}**
- Language family: **{family}**
- Genus: **{genus}**
- Definite articles: **{definite_articles}**
- Indefinite articles: **{indefinite_articles}**
- Number of cases: **{number_of_cases}**
- Order of subject, object, verb: **{word_order}**
- Negative morphemes: **{negative_morphemes}**
- Polar questions: **{polar_questions}**
- Position of negative word wrt SOV: **{position_of_negative_word}**
- Prefixing vs suffixing: **{prefixing_vs_suffixing}**
- Coding of nominal plurality: **{coding_of_nominal_plurality}**
- Grammatical genders: **{grammatical_genders}**

```bibtex
{citations[original_dataset].replace("% ------------------------------------------------------------------------------------------", "")}
```
    """
    
    # Append the markdown row to the output
    markdown_output += markdown_dataset
````

````python
#| eval: false
#| hide
from IPython.display import display, Markdown
````

````python
#| eval: false
#| echo: false

# Display the markdown table
display(Markdown(markdown_output))
````
